```python
x_vals = underform_balde[ : , 0 ]
y_vals = underform_balde[ : , 1 ]
z_vals = underform_balde[ : , 2 ]
x_vals_deform = derform_balde[ : , 0 ]
y_vals_deform = derform_balde[ : , 1 ]
z_vals_deform = derform_balde[ : , 2 ]
fig = plt.figure(figsize = (5 , 5 ))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_vals, y_vals, z_vals, c = 'b', marker='o')
ax.scatter(x_vals_deform, y_vals_deform, z_vals_deform, s=30 , c = 'r', marker='*')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
